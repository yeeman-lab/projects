## Visualizing Wage Differences across Professions and Gender Pay Trends in the UK (1997-2023)

## Purpose of this works

<p> This Python script utilizes the pandas library for data manipulation and matplotlib for visualization. The resulting plots depict gender-based median hourly pay disparities in various occupations across the UK from 1997 to 2023. </p.

## Load in the necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Read CSV files into DataFrame

```python
male = pd.read_csv('male.csv')

print('Dataset for Male:')
male
```

```python
female = pd.read_csv('female.csv')

print('Dataset for Female:')
female
```

## Extract information from DataFrame

```python
# Extract years
year = pd.PeriodIndex(male.columns[1:], freq='Y').year

# Convert the male and female datasets to NumPy arrays
male_rates = male.to_numpy(dtype=None)
female_rates = female.to_numpy(dtype=None)
```

## Graphs data on figure

```python
# Create a new figure with a specified size
fig = plt.figure(figsize=(18, 9))

# Iterate over each occupation for plotting
for i, (male_row, female_row) in enumerate(zip(male_rates, female_rates)):
    sub = fig.add_subplot(2, 3, i+1)
    # Plot median hourly pay for males and females over the years
    sub.plot(year, male_row[1:], label='Male', marker='o')
    sub.plot(year, female_row[1:], label='Female', marker='o')
    # Set plot title, labels, ticks and limits
    sub.set_title(f'Median Hourly Pay\nfor {male_row[0]}')
    sub.set_xlabel('Year')
    sub.set_ylabel('Median Hourly Pay ()')
    sub.set_xticks(year)
    sub.set_ylim(3,30)
    sub.legend()
    sub.grid(True)

plt.tight_layout()
plt.show()
```
